```python
countries = ["Mexico", "Algeria", "Fiji", "Estonia"]

subset = df[df["country_name"].isin(countries)]

years = ["2004", "2005", "2006"]

plt.figure()

for country in countries:
    values = subset[subset["country_name"] == country][years].values.flatten()
    
    plt.scatter(years, values, label=country)

plt.title("GDP Scatter (2004–2006)")
plt.xlabel("Year")
plt.ylabel("GDP")

plt.legend()
plt.show()
```
