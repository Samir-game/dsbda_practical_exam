```python
import pandas as pd

df = pd.read_csv("D:/DATA/aitpune/college study thingy/dsbdal/prac_exam_questions/24-descriptive-stats/Food_Delivery_Times.csv")

df.head()
```

```python
grouped_stats = df.groupby("Traffic_Level")["Delivery_Time_min"].agg(["mean", "median", "min", "max", "std"])

print("Summary Statistics (Delivery Time grouped by Traffic Level):\n")
print(grouped_stats)
```

```python
value_list = df.groupby("Traffic_Level")["Delivery_Time_min"].apply(list)

print("\nList of Delivery Time values for each Traffic Level:\n")
print(value_list)
```

```python
groups = df.groupby("Traffic_Level")["Delivery_Time_min"]

for name, values in groups:
    print(f"\nTraffic Level: {name}")
    
    print("Mean:", values.mean())
    print("Median:", values.median())
    print("Min:", values.min())
    print("Max:", values.max())
    print("Std:", values.std())
    
    print("Values List:", list(values))
```

```python
import seaborn as sns

iris = sns.load_dataset("iris")

iris.head()
```

```python
iris["species"].unique()
```

```python
species_list = iris["species"].unique()
```

```python
for sp in species_list:
    
    print(f"\nStatistics for {sp}:\n")
    
    subset = iris[iris["species"] == sp]
    
    print("Mean:\n", subset.mean(numeric_only=True))
    
    print("\nStandard Deviation:\n", subset.std(numeric_only=True))
    
    print("\nPercentiles:\n", subset.quantile([0.25, 0.5, 0.75], numeric_only=True))
```
