```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Sagar Sharma/OneDrive/Documents/dsbdal assignments/assignment1/train.csv")
df.head()
```

```python
plt.figure()
sns.boxplot(x="Sex", y="Age", hue="Survived", data=df)

plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

plt.show()
```

```python
sns.boxplot(x="Sex", y="Age", hue="Survived", data=df, palette="Set2")
```

```python
plt.figure()

sns.barplot(x="Sex", y="Survived", data=df)

plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate (0-1)")
plt.show()
```

```python
sns.histplot(data=df, x="Age", hue="Survived", bins=30, kde=True)

plt.title("Age Distribution by Survival")
plt.show()
```

```python
survival_by_gender = df.groupby("Sex")["Survived"].mean()

print("Survival Rate by Gender:\n", survival_by_gender)

print("\nObservation:")
print("Females have a higher survival rate compared to males.")
```

```python
df["age_group"] = pd.cut(df["Age"], bins=[0, 12, 18, 40, 60, 80],
                        labels=["Child", "Teen", "Adult", "Middle Age", "Senior"])

survival_by_age = df.groupby("age_group")["Survived"].mean()

print("\nSurvival Rate by Age Group:\n", survival_by_age)

print("\nObservation:")
print("Children have the highest survival rate, indicating priority in rescue.")
```

```python
avg_age = df.groupby("Survived")["Age"].mean()

print("\nAverage Age:\n", avg_age)

print("\nObservation:")
print("Survivors tend to be younger compared to non-survivors.")
```

```python
count_gender = df.groupby(["Sex", "Survived"]).size()

print("\nCount by Gender and Survival:\n", count_gender)

print("\nObservation:")
print("A larger number of male passengers did not survive.")
```
