```python
import pandas as pd

df = pd.read_csv("D:/DATA/aitpune/college study thingy/dsbdal/prac_exam_questions/12-descriptive-stats/winequality-red.csv")

df.head()
```

```python
grouped = df.groupby("quality")["alcohol"].agg(["mean", "median", "min", "max", "std"])

print("Summary Statistics (Alcohol grouped by Quality):\n")
print(grouped)
```

```python
value_list = df.groupby("quality")["alcohol"].apply(list)

print("\nList of alcohol values for each quality:\n")
print(value_list)
```

```python
import seaborn as sns

iris = sns.load_dataset("iris")

iris.head()
```

```python
iris["species"].unique()
```

```python
species_list = iris["species"].unique()
```

```python
for sp in species_list:
    
    print(f"\nStatistics for {sp}:\n")
    
    subset = iris[iris["species"] == sp]
    
    print("Mean:\n", subset.mean(numeric_only=True))
    
    print("\nStandard Deviation:\n", subset.std(numeric_only=True))
    
    print("\nPercentiles:\n", subset.quantile([0.25, 0.5, 0.75], numeric_only=True))
```
