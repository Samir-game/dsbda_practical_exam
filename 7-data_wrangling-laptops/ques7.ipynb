```python
import pandas as pd
import numpy as np

df = pd.read_csv("D:/DATA/aitpune/college study thingy/dsbdal/prac_exam_questions/7-data_wrangling-laptops/laptop_pricing_dataset_mod1.csv")
df.head()
```

```python
print(df.isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df["Weight_kg"] = df["Weight_kg"].astype(str).str.replace("kg", "")
df["Weight_kg"] = pd.to_numeric(df["Weight_kg"], errors='coerce')

df["Screen_Size_cm"] = df["Screen_Size_cm"].astype(str).str.replace("cm", "")
df["Screen_Size_cm"] = pd.to_numeric(df["Screen_Size_cm"], errors='coerce')
```

```python
df["Screen_Size_inch"] = df["Screen_Size_cm"] / 2.54
```

```python
df["Weight_pounds"] = df["Weight_kg"] * 2.20462
```

```python
bins = [0, df["Price"].quantile(0.33), df["Price"].quantile(0.66), df["Price"].max()]
labels = ["Low", "Medium", "High"]

df["Price-binned"] = pd.cut(df["Price"], bins=bins, labels=labels)
```

```python
df = pd.get_dummies(df, columns=["Screen"], drop_first=True)
```
